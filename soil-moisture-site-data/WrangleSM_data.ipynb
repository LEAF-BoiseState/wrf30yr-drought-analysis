```python
import numpy as np
import pandas as pd
import pickle
import glob

#sitedata_dir = '/Users/lejoflores/wrf30yr_drought_analysis/soil-moisture-site-data/sitedata/'
sitedata_dir = './sitedata/'
output_file = 'SRB_site_soil_miosture_data.pkl'

def station_id(file):
    site_id = file.replace(sitedata_dir,'',1)
    site_id = site_id.replace('.csv','',1)
    return(int(site_id))
```

```python
files = sorted(glob.glob(sitedata_dir+'*.csv'),key=station_id)

site_ids = []

sm_dfs = []

for file in files:
    site_id = file.replace(sitedata_dir,'',1)
    site_id = site_id.replace('.csv','',1)
    
    site_ids.append(int(site_id))
    
    print('reading file '+file)
    df = pd.read_csv(file, delimiter=',')
    
    sm_dfs.append(df)
```

```python
sm_data = {'station_id': site_ids,
           'station_data': sm_dfs}

```

```python
with open(output_file, 'wb') as handle:
    pickle.dump(sm_data, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
with open(output_file, 'rb') as handle:
    sm_data_r = pickle.load(handle)
    
sm_data_r['station_data'][0]
```
