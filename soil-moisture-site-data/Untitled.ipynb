```python
import os
import requests
import pandas as pd

site_inventory_file = 'SRB_usda_soimoi_sites.csv'

nsm_url_base='http://nationalsoilmoisture.com/test/VWC_QAQC/'
nsm_ext = '.csv'
```

```python
df = pd.read_csv(site_inventory_file)
df
```

```python
site_id = []
site_files = []

for i in range(df.shape[0]):

    network = df['Network'][i].lower()
    site_id.append(str(df['ID'][i]))
    
    nsm_url = nsm_url_base+network+'/'+site_id[i]+nsm_ext
    
    site_files.append(nsm_url)
    print(nsm_url)
```

```python
for i in range(len(site_files)):
    response = requests.get(site_files[i])
    with open(os.path.join('./sitedata',site_id[i]+nsm_ext), 'wb') as f:
        f.write(response.content)
```
